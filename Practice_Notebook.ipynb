```python
import random

class Flashcard:
  def __init__(self):
    self.fruits = {
      "banana": "yellow",
      "orange": "orange",
      "apple": "red",

      # Add more fruits and colors as needed
    }

  def play(self):
    fruit, color = random.choice(list(self.fruits.items()))
    guess = input(f"What color is a {fruit}? ")
    if guess.lower() == color.lower():
      print("Correct!")
    else:
      print(f"Wrong. A {fruit} is {color}.")

# Create a flashcard object
flashcard = Flashcard()

# Play the game
while True:
  flashcard.play()
  play_again = input("Play again? (y/n): ")
  if play_again.lower() != "y":
    break

print("Thanks for playing!")
```

```python
from collections import Counter

def create_bow(sentences):
  """
  This function takes a list of sentences and returns a list of bag-of-words representations.

  Args:
      sentences: A list of strings representing sentences.

  Returns:
      A list of lists, where each inner list represents the bag-of-words representation
      for the corresponding sentence in the input list.
  """
  # Create a vocabulary of all unique words in the corpus
  vocabulary = set(word for sentence in sentences for word in sentence.lower().split())

  # Function to create a bag-of-words dictionary for a sentence
  def sentence_to_bow(sentence):
    bow_dict = Counter(sentence.lower().split())
    # Fill the dictionary with 0s for missing words in the vocabulary
    for word in vocabulary:
      bow_dict[word] = bow_dict.get(word, 0)
    return list(bow_dict.values())  # Convert dictionary to list

  # Apply the sentence_to_bow function to each sentence
  bow_vectors = [sentence_to_bow(sentence) for sentence in sentences]
  return bow_vectors

# Example usage
sentences = [
  "the quick brown fox jumps over the brown dog",
  "the dog barks loudly",
  "the cat sleeps peacefully"
]

bow_vectors = create_bow(sentences)

print("Sentence 1:", bow_vectors[0])
print("Sentence 2:", bow_vectors[1])
print("Sentence 3:", bow_vectors[2])
```
